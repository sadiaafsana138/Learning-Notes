## 7. MULTIPLE MARKERS (Loop)

## 8. MARKERS FROM DATAFRAME

```python
df = pd.DataFrame({
    'name': ['Dhaka', 'Chittagong', 'Sylhet'],
    'lat':  [23.8103,  22.3569,    24.8949],
    'lon':  [90.4125,  91.7832,    91.8687],
    'pop':  [9000000,  2500000,    500000]
})

m = folium.Map(location=[23.5, 90.5], zoom_start=7)

for _, row in df.iterrows():
    folium.Marker(
        location=[row['lat'], row['lon']],
        popup=f"<b>{row['name']}</b><br>Pop: {row['pop']:,}",
        tooltip=row['name']
    ).add_to(m)

m
```

---
## 9. CIRCLE & CIRCLE MARKER

```python
m = folium.Map(location=[23.8, 90.4], zoom_start=10)

# Circle — radius in metres
folium.Circle(
    location=[23.81, 90.41],
    radius=500,
    color='blue',
    fill=True,
    fill_opacity=0.3
).add_to(m)

# CircleMarker — radius in pixels (stays same size on zoom)
folium.CircleMarker(
    location=[23.75, 90.39],
    radius=10,
    color='red',
    fill=True,
    fill_color='red'
).add_to(m)

m
```

---
## 10. POLYLINE (PATH)

```python
m = folium.Map(location=[23.8, 90.4], zoom_start=10)

points = [[23.81, 90.41], [23.75, 90.39], [23.70, 90.35]]

folium.PolyLine(
    points,
    color='green',
    weight=4,
    opacity=0.8,
    tooltip='Route A'
).add_to(m)

m
```

---
## 11. ANTPATH (Animated Route)

```python
m = folium.Map(location=[23.8, 90.4], zoom_start=10)

points = [[23.81, 90.41], [23.75, 90.39], [23.70, 90.35]]

# Animated flowing dashes — great for routes/flows
AntPath(
    locations=points,
    color='blue',
    weight=5,
    delay=800,       # animation speed in ms
    dash_array=[20, 30]
).add_to(m)

m
```

## 12. POLYGON

```python
m = folium.Map(location=[23.8, 90.4], zoom_start=10)

points = [[23.81, 90.41], [23.75, 90.39], [23.70, 90.35]]

folium.Polygon(
    locations=points,
    color='purple',
    fill=True,
    fill_color='lavender',
    fill_opacity=0.5
).add_to(m)

m
```

## 17. HEATMAP

```python
# heat_data = list of [lat, lon] or [lat, lon, weight]
heat_data = [
    [23.81, 90.41, 1.0],
    [23.75, 90.39, 0.7],
    [23.70, 90.35, 0.5],
    [23.81, 90.42, 0.9]
]

m = folium.Map(location=[23.8, 90.4], zoom_start=10)

HeatMap(
    heat_data,
    radius=15,
    blur=10,
    min_opacity=0.4
).add_to(m)

m
```

## 18. GEOJSON

```python
geo_data = {
    'type': 'FeatureCollection',
    'features': [{
        'type': 'Feature',
        'geometry': {'type': 'Point', 'coordinates': [90.4125, 23.8103]},
        'properties': {'name': 'Dhaka', 'pop': 9000000}
    }]
}

m = folium.Map(location=[23.8, 90.4], zoom_start=10)

folium.GeoJson(
    geo_data,
    tooltip=folium.GeoJsonTooltip(fields=['name', 'pop'], aliases=['City:', 'Population:'])
).add_to(m)

m
```

---
## 19. GEOJSON — Custom Style Function

```python
# Dynamic styling based on feature properties
def style_function(feature):
    pop = feature['properties'].get('pop', 0)
    color = 'red' if pop > 5000000 else 'blue'
    return {
        'color': color,
        'weight': 2,
        'fillOpacity': 0.5
    }

m = folium.Map(location=[23.8, 90.4], zoom_start=7)

folium.GeoJson(
    geo_data,
    style_function=style_function
).add_to(m)

m
```

## 21. LAYER CONTROL (Toggle Layers)

```python
m = folium.Map(location=[23.8, 90.4], zoom_start=10)

# Named feature groups act as toggleable layers
fg1 = folium.FeatureGroup(name='Hospitals')
folium.Marker([23.81, 90.41], tooltip='Hospital A').add_to(fg1)

fg2 = folium.FeatureGroup(name='Schools')
folium.Marker([23.75, 90.39], tooltip='School B').add_to(fg2)

fg1.add_to(m)
fg2.add_to(m)

# Tile layers
folium.TileLayer('OpenStreetMap').add_to(m)
folium.TileLayer('CartoDB positron', name='Light').add_to(m)

folium.LayerControl(collapsed=False).add_to(m)
m
```

## 22. FIT BOUNDS (Auto-zoom to Markers)

```python
locations = [[23.81, 90.41], [23.75, 90.39], [22.33, 91.78]]

m = folium.Map()

for loc in locations:
    folium.Marker(location=loc).add_to(m)

# Auto-zoom to fit all markers — no need to set zoom_start manually
m.fit_bounds(locations)

m
```

## 28. SEARCH PLUGIN (Searchable Markers)

```python
import json

cities = {
    'type': 'FeatureCollection',
    'features': [
        {'type':'Feature','geometry':{'type':'Point','coordinates':[90.4125,23.8103]},'properties':{'name':'Dhaka'}},
        {'type':'Feature','geometry':{'type':'Point','coordinates':[91.7832,22.3569]},'properties':{'name':'Chittagong'}},
        {'type':'Feature','geometry':{'type':'Point','coordinates':[91.8687,24.8949]},'properties':{'name':'Sylhet'}}
    ]
}

m = folium.Map(location=[23.5, 90.5], zoom_start=7)

geo_layer = folium.GeoJson(cities, name='cities').add_to(m)

Search(
    layer=geo_layer,
    search_label='name',
    placeholder='Search city...',
    collapsed=False
).add_to(m)

m
```

## 30. TIMESTAMPED GEOJSON (Animated Time Series)

```python
# Animate points appearing over time
time_data = {
    'type': 'FeatureCollection',
    'features': [
        {
            'type': 'Feature',
            'geometry': {'type': 'Point', 'coordinates': [90.41, 23.81]},
            'properties': {'times': ['2024-01-01'], 'popup': 'Event A'}
        },
        {
            'type': 'Feature',
            'geometry': {'type': 'Point', 'coordinates': [90.39, 23.75]},
            'properties': {'times': ['2024-01-05'], 'popup': 'Event B'}
        },
        {
            'type': 'Feature',
            'geometry': {'type': 'Point', 'coordinates': [91.78, 22.35]},
            'properties': {'times': ['2024-01-10'], 'popup': 'Event C'}
        }
    ]
}

m = folium.Map(location=[23.5, 90.5], zoom_start=7)

TimestampedGeoJson(
    time_data,
    period='P1D',        # interval: P1D=daily, P1M=monthly
    add_last_point=True,
    auto_play=False,
    loop=False
).add_to(m)

m
```
